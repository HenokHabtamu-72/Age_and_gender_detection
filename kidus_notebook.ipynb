```python
import pandas as pd
import numpy as np



data = pd.read_csv(r"..\utkface_metadata.csv")
data.drop(["age_bucket_excel"],axis = "columns", inplace = True)


def match_age_with_range(age: int):
    if age in range(0, 3): return "0-2"
    elif age in range(3, 8): return "4-9"
    elif age in range(8,14): return "8-12"
    elif age in range(14, 24): return "15-20"
    elif age in range(24,35): return "25-32" 
    elif age in range(35,46): return "38-43" 
    elif age in range(46, 65): return "48-53" 
    elif age in range(65, 111): return "60-100"

age = np.array(data.age)
age_bucket_fixed = np.array(map(match_age_with_range, age))
data["age_bucket"] = age_bucket_fixed
data.age_bucket.astype("str")
data.loc[data["age"] == 14]

data.to_csv(r"C:\Users\henok\Data\outputs\results\utkface_metadata_1.csv")
```
